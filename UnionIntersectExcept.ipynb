```python
%sql (SELECT * FROM chinook.track WHERE genreid = 20) union (SELECT * FROM chinook.track WHERE genreid = 18);
```

All entries in track where the genreid is 20 or 18 

```python
%sql (SELECT * FROM chinook.invoice WHERE total < 10) intersect (SELECT * FROM chinook.invoice WHERE total > 5);
```

here we have all entries in the invoices that have a price higher than 5 and lower than 10

```python
%sql (SELECT * FROM chinook.customer WHERE country = 'USA') except (SELECT * FROM chinook.customer WHERE email = 'yahoo.com');
```

All entries of people from USA that don't have a Yahoo mail account.
